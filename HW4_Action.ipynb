```python
#action 1
import numpy as np
import pandas as pd
a = np.array([[0,   0,  0,   1/3,    0,    0], 
              [1/4, 0,  0,   0,      1/2, 0,],
              [0,   1,  0,   1/3,    1/2, 0,],
              [1/4, 0,  0,   0,      0,   1,],
              [1/4, 0,  1,   1/3,    0,   0,],
              [1/4, 0,  0,   0,      0,   0,]])


b = np.array([1/6, 1/6, 1/6, 1/6, 1/6, 1/6])
w = b

def work(a, w):
    print("简化模型模拟结果：")
    for i in range(100):
        w = np.dot(a, w)
        #print(w)
    print("A: %f, B: %f, C: %f, D: %f,E: %f, F: %f" % (w[0],w[1],w[2],w[3],w[4],w[5]))
def random_work(a, w, n):
    print("随机模型模拟结果：")
    d = 0.85
    for i in range(100):
        w = (1-d)/n + d*np.dot(a, w)
        #print(w)
    print("A: %f, B: %f, C: %f, D: %f,E: %f, F: %f" % (w[0],w[1],w[2],w[3],w[4],w[5]))
print("Action 1")    
work(a, w)
random_work(a, w, 6)
print("")

#action 2
from textrank4zh import TextRank4Keyword, TextRank4Sentence
from snownlp import SnowNLP
import jieba

print("Action 2")   
# 打开文件，读取文件
with open("news.txt",encoding='gb2312') as f:  
    text = f.read()  
    
# 输出关键词，设置文本小写，窗口为2
tr4w = TextRank4Keyword()
tr4w.analyze(text=text, lower=True, window=3)
print('textrank4zh算法形成的关键词:')
for item in tr4w.get_keywords(10, word_min_len=2):
    print(item.word, item.weight)
print("")
# 输出重要的句子
tr4s = TextRank4Sentence()
tr4s.analyze(text=text, lower=True, source = 'all_filters')
print('通过textrank4zh算法形成的摘要：')
# 重要性较高的三个句子
for item in tr4s.get_key_sentences(num=3):
	# index是语句在文本中位置，weight表示权重
    print(item.index, item.weight, item.sentence)
print("")

snow = SnowNLP(text)
# 打印关键词
print('SnowNLP算法形成的关键词：')
print(snow.keywords(10))
print(" ")

# TextRank算法
print('SnowNLP算法形成的摘要：')
print(snow.summary(3))
print("运算时间: %s"%snow.sentiments)
    
```
